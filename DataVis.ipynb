```python
df8part2 = pd.read_json('/Users/isaiahhollars/Desktop/q8part2.json')
df8part2.head(15)
```

```python
gym= df8part2.loc[df8part2['Ev_Sport']=='Gymnastics']
swim=df8part2.loc[df8part2['Ev_Sport']=='Swimming']
dive=df8part2.loc[df8part2['Ev_Sport']=='Diving']
figskate=df8part2.loc[df8part2['Ev_Sport']=='Figure Skating']
total = pd.merge(gym,swim,on=['Games_Year'])
total = pd.merge(total,dive,on=['Games_Year'])
total = pd.merge(total,figskate,on=['Games_Year'])
total.head()
```

```python
total.columns = ['gym_age','games_year','gymnastics','swim_age','Swimming','dive_age','Diving','figskate_age','fig skate']
total.head()
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(15,10)
sns.set_context('paper',font_scale=2)
ax = sns.lineplot(data=total,y='gym_age', x='games_year',linewidth=5)
sns.lineplot(data=total,y='swim_age', x='games_year', ax=ax,linewidth=5)
sns.lineplot(data=total,y='dive_age', x='games_year', ax=ax,linewidth=5)
sns.lineplot(data=total,y='figskate_age', x='games_year', ax=ax,linewidth=5)

ax.set_xlabel('Year of Olympics',fontsize=20);
ax.set_ylabel('Avereage Age of Athletes',fontsize=20)
ax.set_title('Average Age of Athletes Over Time by Sport', fontsize=20)

#plt.text(x=1904,y=22.8,s='Gymnastics',fontsize=15)
#plt.text(x=1910,y=21,s='Swimming',fontsize=20)
plt.legend(title='Sport', loc='upper right', \
           labels=['Gymnastics', 'Swimming','Diving','Figure Skating'],\
          prop={'size': 25});
```

```python
fig.savefig("q8part2.png",bbox_inches='tight') 
```
